```python
#Import relevant libraries
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data using pandas as a dataset
df = pd.read_csv('Citywide_Payroll_Data__Fiscal_Year_.csv') 
df.head()
```

```python
#Check summarised information of the dataset
df.info()
```

```python
#Get summary statistics for the dataset
df.describe(include='all')
```

# Assigning appropraite data types

```python
df['Regular Gross Paid'] = df['Regular Gross Paid'].str.replace('$', '').str.strip().astype('float64')
df['Base Salary'] = df['Base Salary'].str.replace('$', '').str.strip().astype('float64')
df['Total OT Paid'] = df['Total OT Paid'].str.replace('$', '').str.strip().astype('float64')
df['Total Other Pay'] = df['Total Other Pay'].str.replace('$', '').str.strip().astype('float64')
df['Agency Start Date'] = pd.to_datetime(df['Agency Start Date'], errors='coerce').dt.date
```

```python
df['Agency Name'] = df['Agency Name'].str.strip()
df['Title Description'] = df['Title Description'].str.strip()
```

```python
#Check null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#Check and drop duplicates
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
df_box = df.select_dtypes(include=['float64'])
# Shuffle rows if needed
#df_box = df_box.sample(frac=1, random_state=1)
```

```python
# Check for Outliers
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxplot(data=df_box[['Base Salary', 'Regular Gross Paid', 'Total OT Paid', 'Total Other Pay']])
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.boxplot(data=df_box[['OT Hours', 'Regular Hours']])

plt.xticks(rotation=45)
plt.show()
```

```python
df_box_cleaned = df_box[['Base Salary', 'Regular Gross Paid']]
threshold = 1.5
for col in df_box_cleaned.columns:
    Q1 = df_box_cleaned[col].quantile(0.25)
    Q3 = df_box_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    df_box_cleaned = df_box_cleaned[(df_box_cleaned[col] >= lower_bound) & (df_box_cleaned[col] <= upper_bound)]
    
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df_box_cleaned[['Base Salary', 'Regular Gross Paid']])
plt.xticks(rotation=45)
plt.show()
```

```python
df_box.corr()
```

```python
# Calculate the correlation matrix
corr_matrix = df_box.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlation Metrics')
plt.show()
```

**How many people do the various city agencies employ, and how much does each department spend on salary in total?**

```python
#Number of employees per city agency
employees_per_city_agency=df['Agency Name'].value_counts().head(15)
employees_per_city_agency.reset_index(name='Employees')
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=employees_per_city_agency.values, y=employees_per_city_agency.index,hue=employees_per_city_agency.index, palette='viridis')
plt.xlabel('Number of Employees')
plt.ylabel('Agency Name')
plt.title('Number of Employees per City Agency')
plt.show()
```

```python
#Total amount paid by agencies as salary
df_salary = df.groupby('Agency Name')['Regular Gross Paid'].sum().reset_index(name='Total paid as salary')
df_salary = df_salary.sort_values(by='Total paid as salary', ascending=False).head(15)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Total paid as salary', y='Agency Name', data=df_salary, hue='Agency Name', palette='plasma')
plt.xlabel('Total Paid as Salary')
plt.ylabel('Agency Name')
plt.title('Total Salary Paid by Top 15 Agencies')
plt.show()
```

```python
# Most numerous job titles in civic government employment
most_numerous_titles = df['Title Description'].value_counts().head(15)
most_numerous_titles
```

```python
# Group by Agency Name and sum the Total OT Paid
overtime_pay_by_agency = df.groupby('Agency Name')['Total OT Paid'].sum().reset_index(name='Total OT Paid')

# Sort the result to find the agencies with the highest overtime pay
overtime_pay_by_agency_sorted = overtime_pay_by_agency.sort_values(by='Total OT Paid', ascending=False)

# Display the top 15 agencies with the highest overtime pay
overtime_pay_by_agency_sorted.head(15)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Total OT Paid', y='Agency Name', hue='Agency Name',data=overtime_pay_by_agency_sorted.head(15), palette='coolwarm')
plt.xlabel('Total OT Paid')
plt.ylabel('Agency Name')
plt.title('Total Overtime Pay by Top 15 Agencies')
plt.show()
```

```python

# Group by Fiscal Year and Agency Name, and sum the Regular Gross Paid
salary_trend = df.groupby(['Fiscal Year', 'Agency Name'])['Regular Gross Paid'].sum().reset_index()

# Identify top agencies based on total Regular Gross Paid across all years
top_agencies = salary_trend.groupby('Agency Name')['Regular Gross Paid'].sum().nlargest(15).index

# Filter the dataset to include only these top agencies
filtered_trend = salary_trend[salary_trend['Agency Name'].isin(top_agencies)]

# Ensure Fiscal Year is sorted properly
filtered_trend = filtered_trend.sort_values(by=['Fiscal Year', 'Regular Gross Paid'], ascending=[True, False])
```

```python
# Plot the trend
plt.figure(figsize=(14, 8))
sns.lineplot(data=filtered_trend, x='Fiscal Year', y='Regular Gross Paid', hue='Agency Name', marker='o')

plt.xlabel('Fiscal Year')
plt.ylabel('Total Regular Gross Paid')
plt.title('Trend of Salary by Fiscal Year by Agency')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
df1= df.groupby('Fiscal Year')['Regular Gross Paid'].sum().reset_index()
```

```python
# Plot the salary trend over fiscal years
plt.figure(figsize=(14, 8))
sns.lineplot(data=df1, x='Fiscal Year', y='Regular Gross Paid', marker='o')

plt.xlabel('Fiscal Year')
plt.ylabel('Total Regular Gross Paid')
plt.title('Salary Trend Over Fiscal Years')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
# Group by Fiscal Year and sum the Total OT Paid
overtime_trend = df.groupby('Fiscal Year')['Total OT Paid'].sum().reset_index()

# Plot the trend
plt.figure(figsize=(14, 8))
sns.lineplot(data=overtime_trend, x='Fiscal Year', y='Total OT Paid', marker='o')

plt.xlabel('Fiscal Year')
plt.ylabel('Total OT Paid')
plt.title('Trend of Overtime Pay by Fiscal Year')
plt.grid(True)
plt.show()
```

```python
# Calculate the percentage of salary growth
df1['Salary Growth (%)'] = df1['Regular Gross Paid'].pct_change() * 100

# Filter the rows for the years 2015, 2016, and 2017
salary_growth = df1[df1['Fiscal Year'].isin(df1['Fiscal Year'].unique())]

# Display the result
salary_growth
```

```python
from matplotlib.colors import LogNorm
sns.barplot(x='Fiscal Year', y='Salary Growth (%)', data=salary_growth)
plt.yscale('log')
plt.xlabel('Fiscal Year')
plt.ylabel('Salary Growth (%)')
plt.title('Salary Growth Percentage by Fiscal Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Regression plot
plt.figure(figsize=(10, 6))
sns.regplot(x='Regular Hours', y='OT Hours', data=df_box, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.xlabel('Regular Hours')
plt.ylabel('OT Hours')
plt.title('Regression Plot of Regular Hours vs OT Hours')
plt.grid(True)
plt.show()
```
